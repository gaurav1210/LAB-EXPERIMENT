```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck constant (J路s)
k = 1.381e-23  # Boltzmann constant (J/K)
c = 3.0e8      # Speed of light (m/s)

# Temperature values for visualization
temperatures = [2.73, 100, 300, 1000]  # Cosmic background, room temp, and higher temps

# Frequency range (Hz)
freq = np.linspace(1e10, 1e14, 1000)  # From 10 GHz to 100 THz

# Planck function definition
def planck_function(frequency, temperature):
    return (2 * h * frequency**3) / (c**2 * (np.exp(h * frequency / (k * temperature)) - 1))

# Visualization of blackbody radiation spectra for different temperatures
plt.figure(figsize=(8, 4))

for T in temperatures:
    spectral_radiance = planck_function(freq, T)
    plt.plot(freq / 1e12, spectral_radiance, label=f"T = {T} K")

# Wien's Displacement Law peak frequencies
peak_frequencies = [2.82 * k * T / h for T in temperatures]

# Marking the peak frequencies
for T, peak_freq in zip(temperatures, peak_frequencies):
    plt.axvline(peak_freq / 1e12, linestyle="--", label=f"Peak for T = {T} K", alpha=0.7)

# Plot customization
plt.title("Blackbody Radiation Spectrum and Wien's Displacement Law", fontsize=12)
plt.xlabel("Frequency (THz)", fontsize = 12)
plt.ylabel("Spectral Radiance (W路m$^{-2}$路Hz$^{-1}$路sr$^{-1}$)", fontsize=12)
plt.legend(fontsize=6)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck's constant (J路s)
k = 1.381e-23  # Boltzmann's constant (J/K)
c = 3.0e8      # Speed of light (m/s)

# Define frequency and temperature
T = 5000  # Temperature in K
frequencies = np.linspace(1e9, 1e15, 1000)  # Frequency range (Hz)

# Planck function
def planck(f, T):
    return (2 * h * f**3 / c**2) / (np.exp(h * f / (k * T)) - 1)

# Rayleigh-Jeans approximation
def rayleigh_jeans(f, T):
    return (2 * f**2 * k * T) / c**2

# Compute values
B_nu_planck = planck(frequencies, T)
B_nu_rj = rayleigh_jeans(frequencies, T)

# Plot Planck vs Rayleigh-Jeans Approximation
plt.figure(figsize=(10, 6))
plt.plot(frequencies / 1e12, B_nu_planck, label="Planck's Law", color="red")
plt.plot(frequencies / 1e12, B_nu_rj, '--', label="Rayleigh-Jeans Approximation", color="blue")

# Labels and title
plt.xlabel("Frequency (THz)")
plt.ylabel("Spectral Radiance (W/m虏/Hz/sr)")
plt.title("Planck vs Rayleigh-Jeans Approximation")
plt.legend()
plt.grid(True)
plt.ylim(0, 1e-14)
plt.show()

# Define parameters for flux density equation
theta_arcmin = 1  # Angular size in arcminutes
Tb = 10  # Brightness temperature in K
wavelengths_cm = np.linspace(0.1, 10, 1000)  # Wavelength in cm

# Compute flux density using the given equation
F_nu = 2.65 * (Tb / 1) * (theta_arcmin / 1)**2 * (wavelengths_cm / 1)**-2

# Plot Flux Density vs Wavelength
plt.figure(figsize=(10, 6))
plt.plot(wavelengths_cm, F_nu, label="Flux Density", color="green")

# Labels and title
plt.xlabel("Wavelength (cm)")
plt.ylabel("Flux Density (Jy)")
plt.title("Flux Density as a Function of Wavelength")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 1.381e-23  # Boltzmann constant (J/K)
c = 3.0e8      # Speed of light (m/s)
arcsec_to_rad = 1 / 206265  # Conversion factor for arcsec to radians
arcmin_to_rad = arcsec_to_rad * 60  # Conversion factor for arcmin to radians

# Function to compute F谓
def compute_flux_density(Tb, theta_arcsec, nu_GHz):
    theta_rad = theta_arcsec * arcsec_to_rad
    omega = np.pi * (theta_rad / 2)**2  # Solid angle
    nu = nu_GHz * 1e9  # Convert GHz to Hz
    F_nu = 2 * k * nu**2 * Tb * omega  # Flux density in SI units (W/m^2/Hz)
    F_nu_Jy = F_nu * 1e26  # Convert to Jansky (1 Jy = 10^-26 W/m^2/Hz)
    return F_nu_Jy

# Jupiter Data
jupiter_Tb = 150  # K
jupiter_theta_arcsec = 40  # arcsec
jupiter_frequencies = [1.4, 115]  # GHz
jupiter_F_nu = [compute_flux_density(jupiter_Tb, jupiter_theta_arcsec, nu) for nu in jupiter_frequencies]

# Orion A Data
orion_Tb = [330, 24]  # K at different frequencies
orion_theta_arcmin = 2.5  # arcmin
orion_frequencies = [4.8, 23]  # GHz
orion_F_nu = [compute_flux_density(Tb, orion_theta_arcmin * 60, nu) for Tb, nu in zip(orion_Tb, orion_frequencies)]

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(["Jupiter (1.4 GHz)", "Jupiter (115 GHz)", "Orion A (4.8 GHz)", "Orion A (23 GHz)"],
        jupiter_F_nu + orion_F_nu, color=['blue', 'blue', 'red', 'red'])

plt.ylabel("Flux Density (Jy)")
plt.title("Computed Flux Density for Jupiter and Orion A")
plt.grid(axis="y", linestyle="--")
plt.show()

# Print values
print(f"Jupiter: F谓 at 1.4 GHz = {jupiter_F_nu[0]:.2f} Jy, F谓 at 115 GHz = {jupiter_F_nu[1]:.2f} Jy")
print(f"Orion A: F谓 at 4.8 GHz = {orion_F_nu[0]:.2f} Jy, F谓 at 23 GHz = {orion_F_nu[1]:.2f} Jy")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_telescope_ray_diagram():
    fig, ax = plt.subplots(figsize=(8, 6))

    # Define telescope components
    primary_mirror_x = [-4, 0, 4]
    primary_mirror_y = [3, 0, 3]
    focal_point = (0, -3)

    # Incoming parallel rays
    for x in np.linspace(-3, 3, 5):
        ax.plot([x, 0], [4, -3], 'b', lw=1.5, label='Light Rays' if x == -3 else "")

    # Primary mirror (parabolic reflector representation)
    ax.plot(primary_mirror_x, primary_mirror_y, 'k', lw=2, label='Primary Mirror')

    # Focal point indication
    ax.plot(focal_point[0], focal_point[1], 'ro', label='Focal Point')
    ax.text(focal_point[0] + 0.1, focal_point[1] - 0.2, "Focal Point", fontsize=10, color='r')

    # CCD pixels (representation)
    for x in np.linspace(-1, 1, 5):
        ax.plot([x, x], [-3.5, -3], 'g', lw=2)
    ax.text(0, -3.8, "CCD Sensor Pixels", fontsize=10, color='g', ha='center')

    # Labels and appearance
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.axhline(0, color='k', lw=0.5, linestyle='dashed')
    ax.set_xlabel("Aperture Width")
    ax.set_ylabel("Light Path")
    ax.set_title("Telescope Ray Diagram with CCD Sensor")
    ax.legend()
    plt.show()

plot_telescope_ray_diagram()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define lens properties
focal_length = -5  # Negative for concave lens
lens_position = 0
ray_start_x = -10  # Starting position of rays

# Generate rays at different heights
ray_heights = np.linspace(-4, 4, 7)  # Rays entering from different heights

plt.figure(figsize=(8, 6))
for h in ray_heights:
    # Ray equation: Using small-angle approximation
    x_vals = np.linspace(ray_start_x, lens_position, 100)
    y_vals = np.ones_like(x_vals) * h  # Parallel rays

    # Compute deviation due to spherical aberration
    delta_x = (h**4) / (8 * focal_length**3)  # Deviation formula
    x_after_lens = np.linspace(lens_position, lens_position - 8, 100)
    y_after_lens = h + (x_after_lens - lens_position) * ((h / focal_length) + delta_x)

    # Plot incoming rays
    plt.plot(x_vals, y_vals, 'b', linestyle="dashed", label="Incoming Rays" if h == ray_heights[0] else "")
    # Plot refracted rays
    plt.plot(x_after_lens, y_after_lens, 'r', label="Refracted Rays" if h == ray_heights[0] else "")

# Draw concave lens
lens_x = [0, 0]
lens_y = [-4.5, 4.5]
plt.plot(lens_x, lens_y, 'k', linewidth=3, label="Concave Lens")

# Formatting the plot
plt.axhline(0, color='gray', linestyle='dotted')
plt.axvline(0, color='gray', linestyle='dotted')
plt.xlim(-10, 6)
plt.ylim(-5, 5)
plt.xlabel("Optical Axis (x)")
plt.ylabel("Ray Height (y)")
plt.title("Spherical Aberration in a Concave Lens")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1.5  # Refractive index of the lens
R1 = -20  # Radius of curvature of the first surface (concave)
R2 = np.inf  # Radius of curvature of the second surface (flat)
f_paraxial = 1 / ((n - 1) * (1 / R1 - 1 / R2))  # Paraxial focal length

# Lens equation with spherical aberration
def lens_equation(y):
    # Higher-order term to simulate spherical aberration
    return f_paraxial + 0.1 * y**2

# Ray tracing for multiple rays
y_values = np.linspace(-10, 10, 20)  # Heights of rays from the optical axis
focal_points = [lens_equation(y) for y in y_values]

# Plotting
plt.figure(figsize=(10, 6))

# Draw the lens
lens_x = [0, 0]
lens_y = [-15, 15]
plt.plot(lens_x, lens_y, 'b-', linewidth=2, label="Concave Lens")

# Draw the rays and their focal points
for y, f in zip(y_values, focal_points):
    plt.plot([-20, 0, f], [y, y, 0], 'r--', linewidth=1)
    plt.scatter(f, 0, color='g')

# Highlight the paraxial focal point
plt.scatter(f_paraxial, 0, color='k', s=100, label="Paraxial Focal Point")

# Add labels and legend
plt.title("Spherical Aberration in a Concave Lens")
plt.xlabel("Optical Axis (x)")
plt.ylabel("Height (y)")
plt.axhline(0, color='k', linestyle=':', linewidth=0.5)  # Optical axis
plt.legend()
plt.grid(True)
plt.show()
```
