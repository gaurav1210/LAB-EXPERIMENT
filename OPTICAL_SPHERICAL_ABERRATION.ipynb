```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1.5
R1 = -20
R2 = np.inf
f_paraxial = 1 / ((n - 1) * (1 / R1 - 1 / R2))

def lens_equation(y):
    return f_paraxial + 0.1 * y**2

# Define the shape of the concave lens
def concave_lens_shape(x):
    return np.sqrt(R1**2 - (x - R1)**2) - R1

# Ray tracing for multiple rays
y_values = np.linspace(-8, 8, 20)
focal_points = [lens_equation(y) for y in y_values]

# Plotting
plt.figure(figsize=(12, 8))

# Draw the concave lens
x_lens = np.linspace(-15, 0, 100)
y_lens = concave_lens_shape(x_lens)
plt.plot(x_lens, y_lens, 'b-', linewidth=2, label="Concave Lens")
plt.plot(x_lens, -y_lens, 'b-', linewidth=2)

# Draw the rays and their focal points
for y, f in zip(y_values, focal_points):
    plt.plot([-20, x_lens[np.argmin(np.abs(y_lens - y))]], [y, y], 'r--', linewidth=1)
    plt.plot([x_lens[np.argmin(np.abs(y_lens - y))], f], [y, 0], 'r--', linewidth=1)
    plt.scatter(f, 0, color='g', s=50)
plt.scatter(f_paraxial, 0, color='k', s=100, label="Paraxial Focal Point")

plt.title("Spherical Aberration in a Concave Lens", fontsize=16)
plt.xlabel("Optical Axis (x)", fontsize=14)
plt.ylabel("Height (y)", fontsize=14)
plt.axhline(0, color='k', linestyle=':', linewidth=0.5)
plt.legend(fontsize=12)
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 100  # Radius of curvature of the mirror
f = R / 2  # Focal length for paraxial rays
h_max = 40  # Maximum height of the ray from the optical axis
num_rays = 10  # Number of rays to plot

# Function to calculate the intersection of reflected rays with the optical axis
def reflected_ray_intersection(h, R):
    # For a spherical mirror, the reflected ray intersects the optical axis at:
    # x = R/2 - h^2 / (2R)
    return R / 2 - h**2 / (2 * R)

# Generate rays
heights = np.linspace(-h_max, h_max, num_rays)  # Rays above and below the optical axis
focal_points = [reflected_ray_intersection(abs(h), R) for h in heights]

# Plotting
plt.figure(figsize=(12, 6))

# Mirror surface (spherical concave mirror facing rightward)
theta = np.linspace(-np.pi/6, np.pi/6, 100)  # Limit theta to show a concave mirror
mirror_x = R * np.cos(theta) - R  # Mirror surface facing rightward
mirror_y = R * np.sin(theta)
plt.plot(mirror_x, mirror_y, 'k-', label='Spherical Concave Mirror')

# Optical axis
plt.axhline(0, color='gray', linestyle='--', label='Optical Axis')

# Plot parallel rays and their reflections
for h in heights:
    # Incident parallel ray (coming from left, parallel to the optical axis)
    plt.plot([-R * 2, 0], [h, h], 'b--', linewidth=1, alpha=0.7)

    # Reflected ray (calculated using geometric optics)
    slope = -h / (R / 2 - h**2 / (2 * R))  # Slope of reflected ray
    plt.plot([0, R * 2], [h, h + slope * (R * 2)], 'r-', linewidth=1, alpha=0.7)

# Mark paraxial focal point
plt.scatter([-f], [0], color='g', s=100, label='Paraxial Focal Point')

# Mark marginal focal points
plt.scatter([-fp for fp in focal_points], np.zeros_like(focal_points), color='purple', s=50, label='Marginal Focal Points')

# Labels and legend
plt.title('Spherical Aberration in a Concave Mirror (Facing Rightward)')
plt.xlabel('Distance along Optical Axis')
plt.ylabel('Height from Optical Axis')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.xlim(-R * 2, R)
plt.ylim(-h_max * 1.5, h_max * 1.5)
plt.show()
```

```python
from pyth.plugins.rtf15.reader import Rtf15Reader
from pyth.plugins.plaintext.writer import PlaintextWriter

def read_rtf_file(file_path):
    # Open the RTF file
    with open(file_path, 'rb') as file:
        # Read the RTF content
        rtf_content = file.read()

        # Parse the RTF content using Rtf15Reader
        document = Rtf15Reader.read(rtf_content)

        # Convert the RTF document to plain text
        plain_text = PlaintextWriter.write(document).getvalue()

        return plain_text

# Example usage
file_path = '/content/BPSDV.rtf'  # Replace with the path to your RTF file
text = read_rtf_file(file_path)
print(text)
```
