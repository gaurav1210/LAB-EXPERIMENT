### **Combined Spectral Plot**
### **Type 1**

```python
# Importing matplotlib library for plotting
import matplotlib.pyplot as plt
# Data from the table
wavelengths = [470, 555, 585, 605, 635, 660, 695, 880, 940]
H1_R = [0.19, 0.08, 0.26, 0.63, 0.50, 0.86, 0.38, 0.85, 0.58]
H2_R = [0.21, 0.23, 0.12, 0.14, 0.07, 0.11, 0.27, 0.58, 0.57]
H3_R = [0.12, 0.17, 0.10, 0.09, 0.06, 0.10, 0.28, 0.56, 0.56]
H4_R = [0.16, 0.22, 0.16, 0.12, 0.09, 0.13, 0.29, 0.51, 0.49]
# Simple line plot
plt.figure(figsize=(8, 4))
plt.plot(wavelengths, H1_R, marker='o', color='orange', label='H1/R')
plt.plot(wavelengths, H2_R, marker='o', color='red', label='H2/R')
plt.plot(wavelengths, H3_R, marker='o', color='pink', label='H3/R')
plt.plot(wavelengths, H4_R, marker='o', color='magenta', label='H4/R')
plt.ylim(0, 1)
# Labeling the X axis
plt.xlabel('Wavelength (nm)')
# Labeling the Y axis
plt.ylabel('Reflectance in 0 to 1 scale')
# Title of the plot
plt.title('Simple Line Plot: H1/R, H2/R, H3/R, H4/R vs Wavelength')
plt.legend()
plt.grid(True)
# Method to show the plot
plt.show()
```

### **Type 2**

```python
# Dashed line plot
plt.figure(figsize=(8, 4))
plt.plot(wavelengths, H1_R, marker='o', linestyle='--', color='orange', label='H1/R')
plt.plot(wavelengths, H2_R, marker='o', linestyle='--', color='red', label='H2/R')
plt.plot(wavelengths, H3_R, marker='o', linestyle='--', color='pink', label='H3/R')
plt.plot(wavelengths, H4_R, marker='o', linestyle='--', color='magenta', label='H4/R')
plt.ylim(0, 1)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance in 0 to 1 scale')
plt.title('Dashed Line Plot: H1/R, H2/R, H3/R, H4/R vs Wavelength')
plt.legend()
plt.grid(True)
plt.show()
```

### **Type 3**

```python
# Filled area plot
plt.figure(figsize=(8, 4))
plt.fill_between(wavelengths, H1_R, color='orange', alpha=0.4, label='H1/R')
plt.fill_between(wavelengths, H2_R, color='red', alpha=0.4, label='H2/R')
plt.fill_between(wavelengths, H3_R, color='pink', alpha=0.4, label='H3/R')
plt.fill_between(wavelengths, H4_R, color='magenta', alpha=0.4, label='H4/R')
plt.ylim(0, 1)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance in 0 to 1 scale')
plt.title('Filled Area Plot: H1/R, H2/R, H3/R, H4/R vs Wavelength')
plt.legend()
plt.grid(True)
plt.show()
```
