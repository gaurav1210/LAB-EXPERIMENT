### **Spectral Plot of Hyperspectral dataset for all Group**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = '/content/2024_10_1_LAB-2_Spectral_Data.csv'
data = pd.read_csv(file_path)

# Step 2: Extract wavelength values
wavelengths = data['Wavelength (nm)']

# Step 3: Plot spectra for all samples
plt.figure(figsize=(23, 12))

# Step 4: Loop through each sample column and plot it
for column in data.columns[1:]:  # Skip the 'Wavelength (nm)' column
    plt.plot(wavelengths, data[column], label=column)

# Step 5: Customize the plot with labels, title, legend, and grid
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.title('Spectra of All Samples Measured by All Groups (FieldSpec Instrument)')
plt.legend(loc='best', fontsize='small', ncol=2)  # Small font size for the legend and two columns
plt.grid(True)

# Step 6: Show the plot
plt.show()
```
